# Do all pre processing on diabetes csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

# Import diabetes.csv data ser using sklearn

```python
df = pd.read_csv('diabetes.csv')
df
```

# Check the distribution of the target

```python
df["Outcome"].value_counts()
```

# Replace zeros with NaN for specific columns

```python
columns_to_replace = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
df[columns_to_replace] = df[columns_to_replace].replace(0,np.nan)
```

# Check for missing values

```python
print(df.isnull().sum())
```

# Fill missing values with median

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df.head()
```

```python
print(df.isnull().sum())
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({'Non Diabetic': 0, 'Diabetic': 1})
df.head()
```

# Visualize Distributions

```python
sns.pairplot(
    df,
    hue="Outcome",        # 0 = Non Diabetic, 1 = Diabetic
    # diag_kind="kde",      # density curve on diagonal
    # palette="Set1"
)

plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
     df.corr(),
    annot=True,        # show correlation values
    fmt=".2f",         # 2 decimal points
    cmap="coolwarm",   # color scheme
    linewidths=0.5,
    cbar=True,
    square=True
)

plt.title("Correlation Heatmap of Diabetes Dataset", fontsize=14)

plt.show()
```

# Divide the data into input and output

```python
x = df.iloc[::,::-1]
x
```

```python
y = df['Outcome']
y
```

# import KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

# Convert all features into Z score

```python
Scalar = StandardScaler()
```

```python
x_StandardScaler = Scalar.fit_transform(x)
```

```python
x_StandardScaler
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_StandardScaler,y,test_size=0.3,random_state=0)
```

# Create KNN Model

```python
model = KNeighborsClassifier(n_neighbors=5,)
```

# Fitting KNN   on dataset

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Display classification_report

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,classification_report
```

```python
y_predict = model.predict(x_test)
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
display = ConfusionMatrixDisplay(cm,display_labels=['a','b'])
```

```python
display.plot()
```

```python
print(classification_report(y_test,y_predict))
```
