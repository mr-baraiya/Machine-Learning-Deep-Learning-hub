# SVR

# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read World bank CSV

```python
df = pd.read_csv("WorldBank.csv")
df
```

# Perform conditional selection to find - Population ages 15-64 (% of total population)

```python
data = df[df['Indicator Name'] == 'Population ages 15-64 (% of total population)']
data
```

# Divide the data into input and output

```python
x = np.arange(1960,2020).reshape(-1,1)
y = data.values[0][4:-1].reshape(-1,1)
```

```python
x
```

```python
y
```

# Plot scatter plot of Population ages 15-64 (% of total population)

```python
plt.scatter(x,y)
plt.show()
```

# Feature Scaling (Mandatory for SVR)**
SVR is highly sensitive to the range of data points. If we don't scale (normalize) the data, the model will fail to find the correct hyperplane.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
sc_y = StandardScaler()

x_scaled = sc_x.fit_transform(x)
y_scaled = sc_y.fit_transform(y)

print("Scaling Complete.")
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y_scaled,test_size=0.3,random_state=0)
```

```python
print(x_train)
```

# Fitting SVR on 3 Different Kernel on dataset

```python
from sklearn.svm import SVR
```

```python
model = SVR()
model.fit(x_train,y_train.ravel())
```

```python
x_grid = np.arange(min(x_test),max(x_test),1)
x_grid = x_grid.reshape(len(x_grid),1)

plt.scatter(x,y,color='red',label='Real Data')

plt.plot(sc_x.inverse_transform(x_grid),sc_y.inverse_transform(model.predict(x_grid).reshape(-1,1)),
        color='blue', label='SVR (RBF)')

plt.title('')
plt.title('')
plt.legend('')
plt.show()
```

```python
model_lin = SVR(kernel='linear')
model_poly = SVR(kernel='poly',degree=3)
model_rbf = SVR(kernel='rbf')
```

```python
model_lin.fit(x_train,y_train.ravel())
model_poly.fit(x_train,y_train.ravel())
model_rbf.fit(x_train,y_train.ravel())
```

```python
plt.figure(figsize=(12,8))

plt.scatter(x, y, color='darkorange', label='Data')

plt.plot(
    sc_x.inverse_transform(x_grid),
    sc_x.inverse_transform(model_lin.predict(x_grid).reshape(-1,1)),
    color='navy',
    label='Linear model'
)

plt.plot(
    sc_x.inverse_transform(x_grid),
    sc_y.inverse_transform(model_poly.predict(x_grid).reshape(-1,1)),
    color='purple',
    label='Polynomial model'
)

plt.plot(
    sc_x.inverse_transform(x_grid),
    sc_y.inverse_transform(model_rbf.predict(x_grid).reshape(-1,1)),
    color='green',
    label='RBF model'
)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('SVR Models Comparison')
plt.legend()
plt.show()
```

# Predict the x_test using 3 Kernel

```python
model_rbf.score(x_train,y_train)
```

```python
model_rbf.score(x_test,y_test)
```
