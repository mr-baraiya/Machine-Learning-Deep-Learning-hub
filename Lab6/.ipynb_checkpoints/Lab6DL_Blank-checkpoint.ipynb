# Decision Tree Classifier

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
data = load_iris()
```

```python
type(data)
```

# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

# Divide the data into input and output

```python
x = data["data"]
```

```python
y = data["target"]
y
```

# Splitting the dataset into the Training set and Test set

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

# Fitting DecisionTreeClassifier   on dataset

```python
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(x_train, y_train)
```

# Display Decision Tree

```python
%whos
```

```python
%matplotlib qt
```

```python
plt.figure(figsize=(20, 10))
tree.plot_tree(
    model,
    filled=True,
    feature_names=data.feature_names,
    class_names=data.target_names,
    rounded=True
)
plt.show()
```

# Predict the x_test 

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
```

# Import diabetes.csv dataset 

```python
df = pd.read_csv('diabetes.csv')
df
```

# Check the distribution of the target

```python
print(df['Outcome'].value_counts())
```

# Replace zeros with NaN for specific columns

```python
columns_to_replace = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
df[columns_to_replace] = df[columns_to_replace].replace(0,np.nan)
```

# Check for missing values

```python
print(df.isnull().sum())
```

# Fill missing values with median

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
print(df.isnull().sum())
```

# Visualize Distributions

```python
df.hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='Outcome', data=df)
plt.title('Distribution of Outcome')
plt.show()
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({'Non Diabetic': 0, 'Diabetic': 1})
```

```python
df.head()
```

# Divide the data into input and output

```python
X = df.drop('Outcome', axis=1)
X
```

```python
Y = df['Outcome']
Y
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

# Create Model

```python
dt_model = DecisionTreeClassifier(random_state=42)
```

# Fitting DecisionTreeClassifier   on dataset

```python
dt_model.fit(X_train, Y_train)
```

# Display Decision Tree

```python
plt.figure(figsize=(25, 15))
tree.plot_tree(
    dt_model,
    filled=True,
    feature_names=X.columns,
    class_names=['No Diabetes', 'Diabetes'],
    rounded=True
)
plt.show()
```

# Predict the x_test 

```python
Y_pred = dt_model.predict(X_test)
Y_pred
```

# Display Training Accuracy

```python
train_accuracy = accuracy_score(Y_train, dt_model.predict(X_train))
print("Training Accuracy:", train_accuracy)
```

# Display Test Accuracy

```python
test_accuracy = accuracy_score(Y_test, Y_pred)
print("Test Accuracy:", test_accuracy)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Diabetes', 'Diabetes'],
            yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
